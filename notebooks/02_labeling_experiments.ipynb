```python
import sys
import pandas as pd
sys.path.append('..')
from training.labeling import create_priority_labels
```

```python
df = pd.read_csv('../data/raw/twcs.csv', nrows=5000)
df = df[df['inbound'] == True].dropna(subset=['text'])
df['text'] = df['text'].str.strip()
```

```python
df['priority'] = df['text'].apply(create_priority_labels)
df['priority'].value_counts()
```

```python
urgent = df[df['priority'] == 'urgent']
print(f"Urgent count: {len(urgent)}")
for text in urgent['text'].sample(10, random_state=42):
    print(text)
    print('---')
```
